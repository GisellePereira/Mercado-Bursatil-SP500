```python
def calcular_roi_sectores(inversion_inicial):
    tickers_sec = ["XLY", "XLV", "XLI", "XLK"]
    data_sect_total = yf.download(tickers_sec, start="2000-01-01", end="2023-02-23")

    for sector in tickers_sec:
        # Obtener el precio de cierre ajustado más reciente
        precio_actual = data_sect_total["Adj Close"].tail(1)[sector]
        # Calcular el precio de cierre ajustado hace 1 año
        precio_hace_un_anio = data_sect_total["Adj Close"].shift(252).dropna()[sector][0]
        # Calcular el ROI para el último año
        ganancia = precio_actual - precio_hace_un_anio
        roi = (ganancia / precio_hace_un_anio) * 100
        # Calcular el valor de la inversión después de 1 año
        valor_final = inversion_inicial * (1 + roi / 100)
        # Imprimir los resultados para cada sector
        print(f"Para el sector {sector}:")
        print(f"Inversión inicial: ${inversion_inicial}")
        print(f"ROI para el último año: {round(roi, 2)}%")
        print(f"Valor final después de 1 año: ${round(valor_final, 2)}\n")
```

```python
calcular_roi_sectores(1000)
```
