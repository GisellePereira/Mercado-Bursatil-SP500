- Buscamos maximizar su retorno de inversión (ROI)

 Se calcula como la ganancia o pérdida en relación con el costo de la inversión original, y depende de varios factores, como el precio de compra y venta de las acciones, las comisiones y los impuestos

```python
#Librerias necesarias
import pandas as pd
#import numpy as np
#import matplotlib.pyplot as plt
#import seaborn as sns
import yfinance as yf
#import plotly.graph_objs as go
from sklearn.linear_model import LinearRegression
import datetime as dt
```

```python
# Descargar datos históricos para rendimiento total
tickers_seclec = ["XLY", "XLP", "XLE", "XLF", "XLV", "XLI", "XLK", "XLU", "XLB", "XLC"]
data_sect_rec = yf.download(tickers_seclec, start="2000-01-01", end="2023-06-01")
```

```python
def calcular_roi_sectores(inversion_inicial):
    tickers_sec = ["XLY", "XLP", "XLE", "XLF", "XLV", "XLI", "XLK", "XLU", "XLB", "XLC"]
    data_sect_total = yf.download(tickers_sec, start="2000-01-01", end="2023-06-01")

    for sector in tickers_sec:
        # Obtener el precio de cierre ajustado más reciente
        precio_actual = data_sect_total["Adj Close"].tail(1)[sector]
        # Calcular el precio de cierre ajustado hace 1 año
        precio_hace_un_anio = data_sect_total["Adj Close"].shift(252).dropna()[sector][0]
        # Calcular el ROI para el último año
        ganancia = precio_actual - precio_hace_un_anio
        roi = (ganancia / precio_hace_un_anio) * 100
        # Calcular el valor de la inversión después de 1 año
        valor_final = inversion_inicial * (1 + roi / 100)
        # Imprimir los resultados para cada sector
        print(f"Para el sector {sector}:")
        print(f"Inversión inicial: ${inversion_inicial}")
        print(f"ROI para el último año: {round(roi, 2)}%")
        print(f"Valor final después de 1 año: ${round(valor_final, 2)}\n")
```

```python
calcular_roi_sectores(1000)
```
